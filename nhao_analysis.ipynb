#### Biểu đồ trung bình giá/m2 theo giấy tờ pháp lý 

```python
df_filtered = df[df["Giấy tờ pháp lý"] != "Đã có sổ"]

df_grouped = df_filtered.groupby(["Quận/Huyện", "Giấy tờ pháp lý"]).size().reset_index(name="Số lượng")

fig = px.bar(
    df_grouped,
    x="Quận/Huyện",
    y="Số lượng",
    color="Giấy tờ pháp lý",
    title="Phân bố giấy tờ pháp lý theo từng quận/huyện ",
    labels={"Số lượng": "Số lượng bất động sản"},
    barmode="group"
)

fig.show()
```

#### Biểu đồ trung bình giá/m2 theo chiều ngang đối với nhà mặt phố

```python
df_street_front = df[df["Loại hình nhà ở"] == "Nhà mặt phố, mặt tiền"]

df_street_front["Nhóm chiều ngang"] = pd.cut(
    df_street_front["Chiều ngang"],
    bins=[0, 3.5, 5, float("inf")],
    labels=["0-3.5m", "3.5-5m", ">5m"]
)

# Tính trung bình giá/m² theo nhóm chiều ngang
df_grouped = df_street_front.groupby("Nhóm chiều ngang",observed=False)["Giá/m²"].mean().reset_index()

fig = px.bar(
    df_grouped,
    x="Nhóm chiều ngang",
    y="Giá/m²",
    title="Trung bình giá/m² theo chiều ngang đối với nhà mặt phố",
    labels={"Giá/m²": "Trung bình giá/m²", "Nhóm chiều ngang": "Chiều ngang"},
)

fig.show()
```

#### Biểu đồ trung bình giá/m2 theo chiều dài và chiều ngang đối với nhà mặt phố và nhà trong ngõ, hẻm

```python
import pandas as pd
import plotly.express as px

df_alley = df[df['Loại hình nhà ở'] == 'Nhà ngõ, hẻm']
df_street = df[df['Loại hình nhà ở'] == 'Nhà mặt phố, mặt tiền']

df_alley['Nhóm chiều ngang'] = pd.cut(
    df_alley["Chiều ngang"],
    bins=[0, 2, 3.5, 5],
    labels=["<2m", "2-3.5m", "3.5-5m"]
)
df_alley['Nhóm chiều dài'] = pd.cut(
    df_alley["Chiều dài"],
    bins=[0, 5, 10, 14],
    labels=["<5m", "5-10m", "10-15m"]
)

df_street['Nhóm chiều ngang'] = pd.cut(
    df_street["Chiều ngang"],
    bins=[0, 3.5, 5, float("inf")],
    labels=["<2m", "2-3.5m", "3.5-5m"]
)
df_street['Nhóm chiều dài'] = pd.cut(
    df_street["Chiều dài"],
    bins=[0, 5, 10, 15],
    labels=["<5m", "5-10m", "10-15m"]
)

mean_price_by_width_alley = df_alley.groupby("Nhóm chiều ngang")["Giá/m²"].mean().reset_index()
mean_price_by_length_alley = df_alley.groupby("Nhóm chiều dài")["Giá/m²"].mean().reset_index()
mean_price_by_width_street = df_street.groupby("Nhóm chiều ngang")["Giá/m²"].mean().reset_index()
mean_price_by_length_street = df_street.groupby("Nhóm chiều dài")["Giá/m²"].mean().reset_index()

mean_price_by_width_alley["Loại nhóm"] = "Chiều ngang (Nhà ngõ, hẻm)"
mean_price_by_length_alley["Loại nhóm"] = "Chiều dài (Nhà ngõ, hẻm)"
mean_price_by_width_street["Loại nhóm"] = "Chiều ngang (Nhà mặt phố)"
mean_price_by_length_street["Loại nhóm"] = "Chiều dài (Nhà mặt phố)"

mean_price_by_width_alley = mean_price_by_width_alley.rename(columns={"Nhóm chiều ngang": "Kích thước"})
mean_price_by_length_alley = mean_price_by_length_alley.rename(columns={"Nhóm chiều dài": "Kích thước"})
mean_price_by_width_street = mean_price_by_width_street.rename(columns={"Nhóm chiều ngang": "Kích thước"})
mean_price_by_length_street = mean_price_by_length_street.rename(columns={"Nhóm chiều dài": "Kích thước"})

mean_price_combined = pd.concat([
    mean_price_by_width_alley,
    mean_price_by_length_alley,
    mean_price_by_width_street,
    mean_price_by_length_street
])

fig = px.line(
    mean_price_combined,
    x="Kích thước",
    y="Giá/m²",
    color="Loại nhóm",
    title="Giá trung bình/m² theo chiều ngang và chiều dài cho các loại hình nhà",
    labels={"Giá/m²": "Giá trung bình (triệu/m²)", "Kích thước": "Kích thước"},
    color_discrete_map={
        "Chiều ngang (Nhà ngõ, hẻm)": "blue",
        "Chiều dài (Nhà ngõ, hẻm)": "blue",
        "Chiều ngang (Nhà mặt phố)": "red",
        "Chiều dài (Nhà mặt phố)": "red"
    }
)

fig.show()
```
